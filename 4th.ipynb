```python
import pandas as pd

# URL of the webpage
url = "https://en.wikipedia.org/wiki/List_of_Delhi_Metro_stations"

# Step 1: Extract all tables from the webpage
try:
    tables = pd.read_html(url)
    print(f"Number of tables found: {len(tables)}")

    # Step 2: Access the third table
    third_table = tables[2]  # Adjust index if needed
    second_column = third_table.iloc[:, 0]  # Get the second column

    # Step 3: Print names in the second column
    print("Names in the second column of the third table:")
    for i, name in enumerate(second_column, start=1):
        print(f"{i}. {name}")

except ValueError as e:
    print(f"Error extracting tables: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
